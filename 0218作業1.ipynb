```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def my_function(x):
  return -x**2

x = np.linspace(-5, 5, 100)
y = my_function(x)
plt.plot(x, y, color='blue', linestyle='-', linewidth=4)

plt.title('2/18 homework', fontsize=20, fontweight='bold', color='red')
plt.xlabel('X bar', fontsize=16, color='green')
plt.ylabel('Y bar', fontsize=16, color='green')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim([-6, 6])
plt.ylim([-26, 1])

plt.show()
```
